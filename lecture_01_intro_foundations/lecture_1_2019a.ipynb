#### Acting Humanly - A. M. Turing

<img src="resources/turing.png" width="600px">
<center>The Turing Test</center>

Turing (1950), Computing Machinery and intelligence:
- Instead of <font color=red>‚ÄúCan machines think?‚Äù</font> ask <font color=red>‚ÄúCan machines do what we (as thinking entities) can do?‚Äù</font>
- An operational test for intelligent behaviour is ‚ÄúThe Imitation Game‚Äù
- Predicted that by the year 2000 a machine might have a 30% chance of fooling a lay person for 5 minutes.
anticipated major arguments against AI, and suggested components of AI, such as knowledge, reasoning, language, understanding, learning. 
- Not a major AI goal (not reproducible, constructive or amenable to mathematical analysis) ‚Äì mostly interesting when intelligent systems interact with humans.

----

#### Thinking humanly - Cognitive science

1960‚Äôs ‚Äúcognitive revolution‚Äù: information-processing psychology (‚Äònatural computation‚Äô) replaced prevailing orthodoxy of behaviourism.

Requires scientific theories of internal activities of the brain:
- What level of abstraction? Knowledge or circuits.
- How to validate:
    - Predicting & testing behaviour of human subjects, top-down = Cognitive Science
    - Direct identification from neurological data, bottom-up = Cognitive Neuroscience.
    
Once a theory of mind becomes precise, enough we can build a computer program to implement it. AI allows us to test psychology hypotheses against experimental data.


----

#### Thinking rationally: ___Laws of Thought___

- Normative (or prescriptive) rather than descriptive

- Roots in Greek philosophy, direct line through mathematics to modern AI

- Problems:
 - Not all intelligent behaviour is mediated by logical deliberation
 - What is the purpose of thinking? What thoughts should I have out of all the thoughts (logical or otherwise) that I could have?
 - Problems with formalisation of informal knowledge ‚Äì especially under uncertainty
 - Problems with computational resources when performing inference in practice

----

#### Acting rationally

Rational behaviour: <font color="blue">Do the right thing</font>.

* The right thing is to maximize the best (expected) outcome given the available information.
* Does not need to involve thinking (e.g. could be purely reflex-based), but any thinking should be in the service of rational action.

Computational limitations make perfect rationality unachievable, so we must have best program given machine resources (**limited/bounded** rationality)

___We will adopt this notion of rationality for designing "intelligent agents" in this course!___

### Why study AI?

**Science & Philosophy**:
- Attempt to understand humans - learn more about ourselves.
- Attempt to understand the general principles for modelling intelligent systems, whether human, animal or artifact.
- Often inspired by research in philosophy, psychology, linguistics, etc., and often leads to new work in those fields. The very difficulty of creating intelligent systems makes us appreciate how complex and subtle the brain is

- Unlike philosophy and psychology, we have a need to develop intelligent entities, i.e., design systems which can do things previously only done by humans or animals.

- Creating human-level intelligence would change civilization, change the way we understand ourselves.

**Regardless, provides an umbrella of tools, techniques, and mindset for solving many  real-world challenges, e.g.**

### Intelligent Agents

<div class="alert alert-info">
<font color=black>How to study and build agents in a systematic and principled manner?</font>
</div>


<div class="alert alert-danger">
<font color=black>
<b>Definition (AIMA Chapter 2.1):</b>
<br>
<br>
An agent is anything that can be viewed as perceiving its environment through sensors and acting upon that environment through actuators (includes humans, robots, chatbots, thermostats).</font>
</div>



<img src="resources/figure21_baseagent2.png" width="500px">
<small><center>AIMA Figure 2.1, a basic agent. Courtesy of Artificial Intelligence: A Modern Approach (Third edition), Stuart Russell and Peter Norvig</center></small>

----

The agent function (as implemented by the agent programme) maps from prior/build-in knowledge, $\pi$, and percepts, $\mathcal{P}$, to actions $\mathcal{A}$, i.e. $f:\mathcal{P},\pi\rightarrow\mathcal{A}$


*__The percepts__*, $\mathcal{P}$:
* Perceptual inputs, percepts, and sequence/history as reported by the sensors.

*__The actuators and actions__*, $\mathcal{A}$:
* What ever means the agent has to influence the environment via its actuators (visual, physical, audio, computer commands,  etc.)

*__The prior knowledge__*, $\mathcal{\pi}$:
* Any hard coded constraints or knowledge about the environment (e.g. if temperature < -40 degrees, I will not do well)

*__The function__*, $f$:
- An abstract, external characterization of the agent by a mathematical function which can be represented by mathematical object such as a look-up table, a continuous or discrete function.
- Internally implemented as a computer program and runs on a physical device

### A framework: A system view of intelligent agents based on rationality

- **<font color=red>P</font>erformance measure**
     - Defines what "good behaviour" is in a specific context.     
     
- **<font color=red>E</font>nvironment**
   - A specification of the physical (or virtual) environment the agent is expected to operate in.

- **<font color=red>A</font>ctuators**
    - The types and physical properties of the actuators available to the agent. Limits what the agent can do.

- **<font color=red>S</font>ensors**
    - The types and physical properties of the sensors available to the agent. Limita what the agent can know about the environment.


<br>

* The aim of the course is to design agents that do a ‚Äògood job‚Äô acting on their environment
 * What is a good job, i.e. what should the performance measure be?
 * Criteria that determine how successful an agent is.
 * No one fixed measure across all tasks, but we insist on an objective measure that an outside authority can establish (i.e. the agent should not itself define what is good.... like humans tend to do).    
 * Better to design performance measures according to what you (as the designer) wants in the environment, rather than how one thinks the agent should behave.

<b>Definition (extended version of AIMA Chapter 2.2):</b>
<br>
<br>
For each possible percept (sequence), $\mathcal{P}$, a <b>rational</b> agent selects an action (sequence) that is expected to maximize its performance measure, given the evidence provided by the percept (sequence) and whatever prior/build-in knowledge the agent has.</font>
</div>

<br>

The goal and success criteria of a agent is measured against a predefined <font color=blue>performance measure</font> . Examples include profit, safely, survival, legal, test score, comfortability, resource consumption, etc.

* **Rational IS NOT clairvoyant nor omniscient **
    - The agent only rarely know the outcome of the action in the environment.  
    - It might have an idea of the outcome based on the percepts and prior knowledge of the environment, and the agents ability to reason based on these, nothing else.
    - E.g. leaving the house in the morning is considered rational even though there is a remote chance of being hit by a falling brick.
   
   
* **Rationality DOES NOT imply success**
    - Being rational therefore does not imply success in solving the task, or if successful any guarantee that it has been solved in any optimal way. I.e. if the performance measure is wrong the agent will typically not solve the actual task by optimizing the measure.     


* **Rationality CAN LEAD to exploration, learning and autonomy**
    - Optimization of the performance measure can lead to exploration and information gathering (e.g. speed, value etc).
    - E.g. not looking before you cross a quiet country road does not make you more rational if you are then hit by a truck! The risk of crossing without looking is too great (loss of performance, i.e. life) - looking helps maximise performance because you are acting to obtain useful information.

### Task and task environment

- <font color=red>P</font>erformance measure
     - Defines what "good behaviour" is in a specific context.     
     
- **<font color=red>E</font>nvironment**
   - A specification of the physical (or virtual) environment the agent is expected to operate in.

- <font color=red>A</font>ctuators
    - The types and physical properties of the actuators available to the agent. Limits what the agnts can do.

- <font color=red>S</font>ensors
    - The types and physical properties of the sensors available to the agent. Limit what the agent can know
about the environment.



----

- __Fully observable vs. partially observable__ (sometimes posed as accessible vs inaccessible)
    - Fully: Access to all relevant information via the sensors
    - Partial: If noisy or broken sensors - or sensors simply do not capture the relevant information.
    - Unobservable: Some times is might be possible to obtain a goal without any observations
- __Deterministic vs stochastic__
    - Deterministic: Next state of the environment is fully captured by current state and action to be carried out.
    - Stochastic: Can not determine the next state based on current state and action due to randomness (or unknowns) in the environment
- __Static vs dynamic__
    - Static: The world never changes.
    - Dynamic: The world changes while we decide what to do, time matters!
    - Semi: The world is the same, but the performance score changes (e.g. time)
- __Discrete vs. Continuous__
    - Discrete: State of the environment are determined among a set of discrete possibilities (e.g. chess), discrete actions (e.g. move left or rigth), discrete percepts (e.g. dirty, not dirty)
    - Continuous: The world has infinitely many states (e.g. temperature), actions are continuous, percepts are continuous (human vision)
- __Single agent vs multiple agents__
    - Single: The agent is alone in the environment
    - Multi: There are multi agents, either cooperating or competing against each other to fulfill individual or group goals.
- __Episodic vs sequential__
    - Episodic: Single actions based on current percept only
    - Sequential: Current action influences all future decisions
- __Known vs. unknown__
    - Do you know the rules of game (e.g. gravity) or how the buttons do befor eyou try them out?

#### Goal-based

<img src="resources/figure213_goal.png" width="600px">
<small><center>Image: AIMA 2.13 - Goal-based agent. Figure courtesy of Stuart Russell and Peter Norvig, Artificial Intelligence: A Modern Approach</center></small>

<br>

* A specific goal, which can be obtained in one or typically more steps. 
     * E.g. our taxi driver needs to go to Inverness from Glasgow and at any junction he can go (left, right, straight, turn), what should he do ?
* Involves questions such as: "What will happen if I do this - and will it get me closer to the goal". This is not the case in reflex agents: they map directly from percepts to actions.
* Solved by search and planning techniques (which we will consider soon) .

----

#### Utility-based

<img src="resources/figure214_utility.png" width="600px">
<small><center>Image: AIMA 2.14 - Utility-based agent. Figure courtesy of Stuart Russell and Peter Norvig, Artificial Intelligence: A Modern Approach</center></small>

<br>
<br>
<br>

* Still a goal, but be "happy" when you have reached it.
* When there are two ways to obtain the goal, utility provides a tradeoff.
* Potential to cope with uncertainty and partial observability.
* Solved by formulating a number of utility functions, and using search/planning and continuous optimization methods to optimize the expected utility.

#### ... they can all learn

<img src="resources/figure215_learning.png" width="600px">
<small><center>Image: AIMA 2.15 - Learning agent. Figure courtesy of Stuart Russell and Peter Norvig, Artificial Intelligence: A Modern Approach</center></small>


<br>

* Nature (innate qualities) vs. nurture (personal experience).
* If the agent is only dependent on built-in knowledge it lacks autonomy.
* A system is autonomous to the extent that its behaviour is determined by its own experience. Just as animals are born with reflexes and instincts to survive to the point they can learn for themselves, the agent will need some pre-programmed knowledge from the designer.


.... perhaps a "child agent" that can learn is the way to obtain general AI / human level AI ?
